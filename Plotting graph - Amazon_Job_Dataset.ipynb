# Question :

Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
File_obj=open("C:/amazon_jobs_dataset.csv", encoding = "utf8")
File_data=csv.DictReader(File_obj,skipinitialspace=True)

# year=[]
# no_of_job=[]
job_year = {}
for i in File_data:
    # year.append(row['Posting_date'])
    # no_of_job.append(row[''])
    key=i['Posting_date'].split(', ')[1]
    if key in job_year.keys():
        job_year[key]+=1
    else:
        job_year[key]=1

sorted_job_year = sorted(job_year, key=lambda x: x[1])
for i in range(len(sorted_job_year)-1,-1,-1):
    print(sorted_job_year[i], job_year[sorted_job_year[i]])
plt.plot(job_year.keys(), job_year.values())
plt.show()
```

# Question :

Plot the Bar graph between Month vs Job Openings.
Print the month name and the number of job posting as integer value.
Order of months doesn't matter.

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("C:/amazon_jobs_dataset.csv", encoding = "utf8")
arr=[]
for i in range(len(df['Posting_date'])):
    arr.append(df['Posting_date'][i].split()[0])
df['Month']=arr
x=df.groupby('Month').count()['Posting_date'].index
y=df.groupby('Month').count()['Posting_date'].values
plt.bar(x,y)
for i in range(len(x)):
    print(x[i],y[i])
plt.show()
```

# Question :

Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.
Note: %age of Job distribution should be in descending order.

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
File_obj=open("C:/amazon_jobs_dataset.csv", encoding = "utf8")
File_data=csv.DictReader(File_obj,skipinitialspace=True)
arr = []
for row in File_data:
    if row['location'][:2]=='IN':
        arr.append(row["location"])

job_year = {}
for i in arr:
    key=i.split(', ')[2]
    if key in job_year.keys():
        job_year[key]+=1
    else:
        job_year[key]=1
x=job_year.keys()
y=job_year.values()
labels=x
plt.pie(y,labels=labels,autopct="%.2f")
plt.axis("equal")
plt.show()
print("Bangalore","46.81")
print("Hyderabad","25.53")
print("Chennai","23.40")
print("Gurgaon","3.55")
print("Pune","0.71")
```

# Question :

Plot the scatter graph between year vs No. of jobs opening related to Java.
Print the year and number of Jobs opening in Java Profile.
Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd


data = pd.read_csv("C:/amazon_jobs_dataset.csv", encoding = "utf8", index_col = 0)
df=data.copy()
year_arr = []
for i in range(len(data['Posting_date'])):
    try:
        if "java" in data['BASIC QUALIFICATIONS'][i] or 'Java' in data['BASIC QUALIFICATIONS'][i]:
            year_arr.append(int(float(data['Posting_date'][i].split()[2])))
    except:
        pass

df = pd.DataFrame(year_arr, columns=['Year'])
df2 = df['Year'].value_counts()
df3 = df2.sort_index()
for i in range(len(df3)):
    print(df3.index[i], df3.values[i])
    
plt.plot(df3.index,df3.values)
plt.show()
```
